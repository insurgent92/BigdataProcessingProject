## 1. Import pakages

```python
import keras
from keras.callbacks import ModelCheckpoint
from keras import models
from keras import layers
keras.__version__
```

## 2.1 Set model 1
\# of hidden layers : 1

```python
!mkdir drive/colab/BigData/BigdataProcessingProject/Assignment1/models
!ls drive/colab/BigData/BigdataProcessingProject/Assignment1
```

```python

network1 = models.Sequential()

network1.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network1.add(layers.Dense(10, activation='softmax', input_shape=(512,)))
network1.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

network1.save('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_depth_1(no_trained).h5')

```

## 2.2 Set model 2
\# of hidden layers : 3

```python
network2 = models.Sequential()
network2.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network2.add(layers.Dense(512, activation='relu', input_shape=(512,)))
network2.add(layers.Dense(512, activation='relu', input_shape=(512,)))
network2.add(layers.Dense(10, activation='softmax'))

network2.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

network2.save('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_depth_2(no_trained).h5')
```
