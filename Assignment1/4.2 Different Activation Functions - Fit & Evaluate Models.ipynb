## 1. Import pakages

```python
import keras
from keras.callbacks import ModelCheckpoint
from keras import models
from keras import layers
import time
keras.__version__

```

## 2. Import MNIST

```python
from keras.datasets import mnist
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

## 3. Import Models

```python
from keras.models import load_model

network1 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_1(no_trained).h5')
network2 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_2(no_trained).h5')
network3 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_3(no_trained).h5')
network4 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_4(no_trained).h5')
network5 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_5(no_trained).h5')
network6 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_6(no_trained).h5')
network7 = load_model('drive/colab/BigData/BigdataProcessingProject/Assignment1/models/mnist_activation_7(no_trained).h5')
```

```python
model_path1 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model1/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'
model_path2 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model2/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'
model_path3 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model3/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'
model_path4 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model4/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'
model_path5 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model5/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'
model_path6 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model6/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'
model_path7 = 'drive/colab/BigData/BigdataProcessingProject/Assignment1/checkpoint/model7/' + 'loss{epoch:02d}-{val_loss:.4f}.hdf5'

cb_checkpoint1 = ModelCheckpoint(filepath=model_path1, monitor='val_loss', verbose=1, save_best_only=True)
cb_checkpoint2 = ModelCheckpoint(filepath=model_path2, monitor='val_loss', verbose=1, save_best_only=True)
cb_checkpoint3 = ModelCheckpoint(filepath=model_path3, monitor='val_loss', verbose=1, save_best_only=True)
cb_checkpoint4 = ModelCheckpoint(filepath=model_path4, monitor='val_loss', verbose=1, save_best_only=True)
cb_checkpoint5 = ModelCheckpoint(filepath=model_path5, monitor='val_loss', verbose=1, save_best_only=True)
cb_checkpoint6 = ModelCheckpoint(filepath=model_path6, monitor='val_loss', verbose=1, save_best_only=True)
cb_checkpoint7 = ModelCheckpoint(filepath=model_path7, monitor='val_loss', verbose=1, save_best_only=True)
```

## 4. Fit

```python
start_time = time.time() 
result1 = network1.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint1])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
start_time = time.time() 
result2 = network2.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint2])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
start_time = time.time() 
result3 = network3.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint3])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
start_time = time.time() 
result4 = network4.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint4])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
start_time = time.time() 
result5 = network5.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint5])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
result6 = network6.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint6])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
result7 = network7.fit(train_images, train_labels, epochs= 10, batch_size=512, validation_split = 0.1, callbacks = [cb_checkpoint7])
print("start_time", start_time)
print("--- %s seconds ---" %(time.time() - start_time))
```

```python
n1_test_loss, n1_test_acc = network1.evaluate(test_images, test_labels)
print('n1_test_acc:', n1_test_acc)
print('n1_test_loss:', n1_test_loss)

n2_test_loss, n2_test_acc = network2.evaluate(test_images, test_labels)
print('n2_test_acc:', n2_test_acc)
print('n2_test_loss:', n2_test_loss)

n3_test_loss, n3_test_acc = network3.evaluate(test_images, test_labels)
print('n3_test_acc:', n3_test_acc)
print('n3_test_loss:', n3_test_loss)

n4_test_loss, n4_test_acc = network4.evaluate(test_images, test_labels)
print('n4_test_acc:', n4_test_acc)
print('n4_test_loss:', n4_test_loss)

n5_test_loss, n5_test_acc = network5.evaluate(test_images, test_labels)
print('n5_test_acc:', n5_test_acc)
print('n5_test_loss:', n5_test_loss)

n6_test_loss, n6_test_acc = network6.evaluate(test_images, test_labels)
print('n6_test_acc:', n6_test_acc)
print('n6_test_loss:', n6_test_loss)

n7_test_loss, n7_test_acc = network7.evaluate(test_images, test_labels)
print('n7_test_acc:', n7_test_acc)
print('n7_test_loss:', n7_test_loss)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def flotResult(data1,data2,data3,data4,data5):
    plt.figure(figsize=(5,5))
    plt.plot(data1,'r', label='Depth_1')
    plt.plot(data2,'g', label='Depth_3')
    plt.plot(data3,'b', label='Depth_5')
    plt.plot(data4,'c', label='Depth_7')
    plt.plot(data5,'m', label='Depth_19')
    
    plt.legend(loc='upper right')
    plt.show()
    
```

```python
def flotResult2(title,data1,data2):
    plt.figure(figsize=(5,5))
    plt.title(title)
    plt.plot(data1,'r', label='train_loss')
    plt.plot(data2,'g', label='valid_loss')
    
    plt.legend(loc='upper right')
    plt.show()
    
```

```python
flotResult2('softplus', result1.history['loss'], result1.history['val_loss'])
flotResult2('softsign', result2.history['loss'], result2.history['val_loss'])
flotResult2('relu', result3.history['loss'], result3.history['val_loss'])
flotResult2('tanh', result4.history['loss'], result4.history['val_loss'])
flotResult2('sigmoid', result5.history['loss'], result5.history['val_loss'])
flotResult2('hard_sigmoid', result6.history['loss'], result6.history['val_loss'])
flotResult2('linear', result7.history['loss'], result7.history['val_loss'])
```
